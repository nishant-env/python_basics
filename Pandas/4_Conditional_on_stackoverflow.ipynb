```python
import pandas as pd
```

```python
df = pd.read_csv("Data/survey_results_public.csv", index_col="Respondent")
df_schema = pd.read_csv("Data/survey_results_schema.csv", index_col="Column")
df.columns
```

```python
# filtering out the data that makes over 80000 per year
df_schema.loc["ConvertedComp", "QuestionText"]
```

```python
filt = df["ConvertedComp"] >=80000
df_more80 = df.loc[filt, ["Country", "LanguageWorkedWith", "ConvertedComp"]]
df_more80
```

```python
# only the people who are from india, finland or netharlands
country_list = ["India", "Finland", "Netherlands"]
filt2 = df_more80["Country"].isin(country_list)
df_more80_count = df_more80.loc[filt2]
```

```python
df_more80_count
```

```python
## using string function
# filerting out rows for people that uses python
filt_python = df_more80_count['LanguageWorkedWith'].str.contains('Python', na=False)
df_more80_count_pyth = df_more80_count.loc[filt_python]
df_more80_count_pyth
```
