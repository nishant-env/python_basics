```python
import pandas as pd
df = pd.read_csv('Data/ETH_1h.csv')
```

```python
df.dtypes
```

```python
### here the date is shown as object, converting to datetype
# cannot convert as it cannot parse the format
df['Date'] = pd.to_datetime('Date')
```

```python
df
```

```python
## converting into datetime by giving the date format
df["Date"] = pd.to_datetime(df["Date"], format='%Y-%m-%d %I-%p')
```

```python
df
```

```python
### this could directly be done during import
from datetime import datetime
format_date = lambda x:  datetime.strptime(x,'%Y-%m-%d %I-%p')
df_1 = pd.read_csv("Data/ETH_1h.csv", parse_dates=["Date"], date_parser=format_date)


# in newer pandas we can directly pass date_format argument
```

```python
df_1
```

```python
## we can now use different datetime functions with this column
# df["Date"].dt.dayofweek   # - shows in number
df["DayInWeek"] = df["Date"].dt.day_name()
df
```

```python
## similarly many other methods
df["Date"].min()
```

```python
df["Date"].max()
```

```python
# we can also add or subtract the dates
df["Date"].max() - df["Date"].min()
```

```python
## setting the date as index
df.set_index("Date", inplace=True)
df
```

```python
# now we can use this to filter the data according to the dates
df.loc['2017']
```

```python
## for 2017 Nov and Dec data, we can use slicing as
df.loc['2017-11': '2017-12']
```

```python
# for 2017 december daily highest for high column (resampling daywise)
df.loc['2017-12', "High"].resample('D').max()
```

```python
#resampling different columns daywise, but different aggregate function
df.loc['2017-12'].resample("D").agg({"Open": min, "Close" : max})
```
