By default index for a row is a numbered list that is generated by default. If we want to change the index to some column that will uniquely identify our rows, we can do it in the pandas.

```python
people={
    "first": ["Nishant", "Ram", "Shyam"],
    "last" : ["Neupane", "Sitaula", "Dahal"],
    "email" : ["nishant.neupane@gmail.com","ram.sitaula@gmail.com", "shyam.dahal@gmail.com"]
}
import pandas as pd
df = pd.DataFrame(people)
```

```python
df 
```

```python
# here we can see that the default index is 0,1,2. We can set the index to email by 
df.set_index('email', inplace=True)
df
```

```python
## now while using loc, we can specify the email instead of the integers for filtering
df.loc['nishant.neupane@gmail.com', 'first']
```

```python
## going back to the integer index,
df.reset_index(inplace=True)
```
