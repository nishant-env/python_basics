## How many developers know python from india ?

```python
### First using the filters
filt = df.loc[df["Country"]=="India"]["LanguageWorkedWith"].str.contains('Python').sum()
```

```python
filt
```

```python
# Using group by object
#def contains_text
country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains('Python').sum())
```

```python
### or this can be done with a custom function
def worked_python(grp):
    return grp.str.contains("Python").sum()
country_grp["LanguageWorkedWith"].apply(worked_python)
```

## Getting percentage of people who knows python

```python
total_respondents = country_grp["LanguageWorkedWith"].count()
total_respondents
```

```python
worked_with_python = country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
worked_with_python
```

```python
df_comb = pd.concat([total_respondents, worked_with_python], axis="columns")
```

```python
df_comb
```

```python
df_comb.columns= ["Total Respondents", "Worked With Python"]
df_comb
```

```python
df_comb["Python Percent"]= (df_comb["Worked With Python"]/df_comb["Total Respondents"])*100
```

```python
df_comb.sort_values(by="Python Percent", ascending=False)
```
