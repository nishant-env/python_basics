```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
df = pd.DataFrame(people)
df
```

```python
### cleaning row that contains NAN or None

#df.dropna() 
#equivalent to 
df.dropna(axis="index", how="any")   # here index axis mean rows, and any instructs to drop row if any value for 
# that row is null 
```

```python
# to drop row only if all values are missing 
df.dropna(axis="index", how="all")
```

```python
# to create a more complex rule for dropping, subset can be used
# for example, to drop a row if either of email or lastname is missing

df.dropna(axis="index", how="any", subset=["email", "last"])
```

```python
# drop row only if both the email and lastname are missing
df.dropna(axis="index" ,how="all", subset=["email","last"])
```

## similarly, the rules can be applied to drop columns simply by putting axis="columns"


# Here the custom missing values can be changed to a numpy NaN 

```python
df.replace({"NA": np.nan, "Missing": None}, inplace=True)
df
```

```python
### null values can be filled with other values too, eg: filling with 0 
df.fillna(0)
```
